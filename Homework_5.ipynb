wcss is 'kmeans.inertia_'. The KMeans algorithm clusters data by trying to separate samples in n groups of equal variance, minimizing a criterion known as inertia which is the sum of squared error for each cluster.  Inertia can be recognized as a measure of how internally coherent clusters are.

## Problem 1 Solution

```python
#loading the iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#training and getting KMeans data
from sklearn.cluster import KMeans
from sklearn.model_selection import cross_val_score
import numpy as np

k_range = range(1, 11)
scores = {}
for k in k_range:
  kmeans = KMeans(n_clusters=k, random_state=11)
  kmeans.fit(iris.data)
  scores[k] = kmeans.inertia_
print(scores)
```

```python
#Visualizing the KMeans scores
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(k_range, scores.values(),  color='blue')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')

plt.show()
```

With reference to "The Elbow Method" graph above, the best K value for KMeans with the Iris dataset would be K = 3
